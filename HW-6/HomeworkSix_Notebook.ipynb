## Explain the linearity or nonlinearity of the decision boundary, and explain the difference in decision regions for the various values of γ. 
### In 1] (b), We are using Radial Basis Function kernel. Hence, the decision boundary we get in the un-transformed space (input X - space) is non-linear, because the RBF kernel makes a non-linear transformation of the input space to higher dimensions and in this U - space, the data will be linearly separable and we can fit a linear boundary in the U- space.

### We had fixed the Slack Varible C to be 0.01 and we altered the values of gamma (1, 3, 10, 50). In an SVM model with an RBF kernel, the gamma value determines the extent of influence that a single training point has on the classification of other points. A small gamma value corresponds to a large similarity radius, meaning that many points are grouped together and considered similar. In contrast, a large gamma value results in a small similarity radius, implying that only points that are very close to each other are grouped together and considered similar. Therefore, low gamma values tend to result in a smoother decision boundary with larger margins, while high gamma values tend to result in a more complex decision boundary that is highly sensitive to the training data.
