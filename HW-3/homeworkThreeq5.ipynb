## 5. (b) b] Plot ReLu(x) function with the softmax(x) vs x. Discuss the change in the criterion function when this softmax function is used.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
def relu(x):
    rel_x = np.maximum(0,x)
    return rel_x
    
```

```python
def softPlus(x):
    splus_x = np.log(1 + np.exp(x))
    return splus_x 
```

```python
def plotRelu(X):
    ax = plt.axes()

    ax.plot(X, relu(X), color = "#85586F") 
    ax.set_title("ReLU(X) Vs. X")
    ax.set_ylabel("ReLU(X)")
    ax.set_xlabel("X")

    plt.show()
```

```python
def plotSoftPlus(X):
    ax = plt.axes()

    ax.plot(X, softPlus(X), color = "#85586F") 
    ax.set_title("softPlus(X) Vs. X")
    ax.set_ylabel("softPlus(X)")
    ax.set_xlabel("X")

    plt.show()
```

```python
X1 = np.linspace(-10, 10, 20)
X1
```

```python
X2 = np.linspace(-10, 10, 20)
X2
```

```python
relu(X1)
```

```python
softPlus(X2)
```

```python
plotRelu(X1)
```

```python
plotSoftPlus(X2)
```
