### Instantiating objects for <strong> <em> NearestMeansClassifier() class </strong> </em> one for <strong> <em> train_data_1 </strong> </em>  and one for <strong> <em> test_data_1 </strong> </em>.

```python
nmc_1 = nearestMeansClassifier.NearestMeansClassifier()
```

### Generating class_means, Plotting the Decision Boundary and Decision Regions, Classsifying the data-points and Calculating CER & Accuracy for training dataset of dataset - 1.

```python
X_train_1, n_train_1, T_train_1 = nmc_1.generateData(data=train_data_1)

class_means_1 = nmc_1.calculateClassMeans(X=X_train_1)

print("---------------------------------------------------")
print(f"  Shape of sample_means: {class_means_1.shape}")
print(f"  Sample Means: ")
print(class_means_1)
print("---------------------------------------------------")

nmc_1.plotDecisionBoundaries(training=X_train_1, label_train=T_train_1, sample_mean=class_means_1, fsize=(18,18))

Y_hat_train_1 = nmc_1.classify(X=X_train_1, sample_means=class_means_1)

CER_train_1 = nmc_1.calculateCER(T=T_train_1, Y_hat=Y_hat_train_1, percentageFlag=True)

accuracy_train_1 = nmc_1.calculateAccuracy(T=T_train_1, Y_hat=Y_hat_train_1, percentageFlag=True)

print(f"Classification Error Rate for the training set of dataset-1 is: {CER_train_1}%")
print(f"Accuracy for the training set of dataset-1 is: {accuracy_train_1}%")
```

### Plotting the Decision Boundary and Decision Regions, Classsifying the data-points and Calculating CER & Accuracy for testing dataset of dataset - 1.

```python
X_test_1, n_test_1, T_test_1 = nmc_1.transformTestData(test_data=test_data_1)

nmc_1.plotDecisionBoundaries(training=X_test_1, label_train=T_test_1, sample_mean=class_means_1, fsize=(10, 10))

Y_hat_test_1 = nmc_1.classify(X=X_test_1, sample_means=class_means_1)

CER_test_1 = nmc_1.calculateCER(T=T_test_1, Y_hat=Y_hat_test_1, percentageFlag=True)

accuracy_test_1 = nmc_1.calculateAccuracy(T=T_test_1, Y_hat=Y_hat_test_1, percentageFlag=True)

print(f"Classification Error Rate for the testing set of dataset-1 is: {CER_test_1}%")
print(f"Accuracy for the testing set of dataset-1 is: {accuracy_test_1}%")
```

## d] Feature Transformation: Feature reduction by Projection.

### 1] Feature Transformation for DATASET -> 1

### Transforming train_data_1 by performing projection on 40 vectors produced by brute-force approach.

```python
nmc_1_std_r = nearestMeansClassifier.NearestMeansClassifier()

X_train_1, n_train_1, T_train_1 = nmc_1_std_r.generateData(data=train_data_1, printFlag=True)

X_train_1_standardized = nmc_1_std_r.standardizeTrainData(X_train=X_train_1, printFlag=True)

m_1, CER_History_1, accuracy_History_1 = nmc_1_std_r.featureTransformationTrain(X_train_std=X_train_1_standardized, 
                                                                          n_train=n_train_1, 
                                                                          T_train=T_train_1, 
                                                                          datasetName="train_dataset_1",
                                                                          printFlag=False)
```

### Printing m_star, rm_star and class_means_rm_star for train_data_1

```python
print(nmc_1_std_r.m_star)
print(nmc_1_std_r.rm_star)
print(nmc_1_std_r.class_means_rm_star)

print(f"The least CER Produced at index {nmc_1_std_r.m_star} is {CER_History_1[nmc_1_std_r.m_star]}%")
print(f"The Highest Accuracy produced at index {nmc_1_std_r.m_star} is {accuracy_History_1[nmc_1_std_r.m_star]}%")
```

### Plotting the CER vs. # of vectors(m) and Accuracy vs. # of vectors(m) for train_data_1

```python
pg.plotGraphCER(m=m_1, CER_History=CER_History_1, datasetName="Train_Set_1", leastCER=CER_History_1[nmc_1_std_r.m_star])
```

```python
pg.plotGraphAcc(m=m_1, accuracy_History=accuracy_History_1, datasetName="Train_Set_1", highestAcc=accuracy_History_1[nmc_1_std_r.m_star])
```

### Transforming test_data_1 by performing projection on rm_star and calculating CER & Accuracy using the class_means_rm_star.

```python
X_test_1, n_test_1, T_test_1 = nmc_1_std_r.transformTestData(test_data=test_data_1)

X_test_1_standardized = nmc_1_std_r.standardizeTestData(X_test=X_test_1)

CER_test_reduced_1, accuracy_test_reduced_1, X_test_reduced_1 = nmc_1_std_r.featureTransformationTest(X_test_std=X_test_1_standardized, 
                                                                                                    n_test=n_test_1, 
                                                                                                    T_test=T_test_1)

print(f"The Classification Error Rate of projected test data of Dataset-1 is: {CER_test_reduced_1}")
print(f"The Accuracy of projected test data of Dataset-1 is: {accuracy_test_reduced_1}")
```

### Plotting the Projected Data Points, class_means_rm_star, Decision Boundary and Decision Regions for test_data_1

```python
nmc_1_std_r.plotTranformedTestData(X_test_reduced_1, T_test_1)
```
