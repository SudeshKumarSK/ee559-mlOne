### Generating class_means, Plotting the Decision Boundary and Decision Regions, Classsifying the data-points and Calculating CER & Accuracy for training dataset of dataset - 1.

```python
X_train_1, n_train_1, T_train_1 = nmc_1.generateData(data=train_data_1)

class_means_1 = nmc_1.calculateClassMeans(X=X_train_1)

print("---------------------------------------------------")
print(f"  Shape of sample_means: {class_means_1.shape}")
print(f"  Sample Means: ")
print(class_means_1)
print("---------------------------------------------------")

nmc_1.plotDecisionBoundaries(training=X_train_1, label_train=T_train_1, sample_mean=class_means_1, fsize=(10,5))

Y_hat_train_1 = nmc_1.classify(X=X_train_1)

CER_train_1 = nmc_1.calculateCER(T=T_train_1, Y_hat=Y_hat_train_1, percentageFlag=True)

accuracy_train_1 = nmc_1.calculateAccuracy(T=T_train_1, Y_hat=Y_hat_train_1, percentageFlag=True)

print(f"Classification Error Rate for the training set of dataset-1 is: {CER_train_1}")
print(f"Accuracy for the training set of dataset-1 is: {accuracy_train_1}")
```

### Plotting the Decision Boundary and Decision Regions, Classsifying the data-points and Calculating CER & Accuracy for testing dataset of dataset - 1.

```python
X_test_1, n_test_1, T_test_1 = nmc_1.transformTestData(test_data=test_data_1)

nmc_1.plotDecisionBoundaries(training=X_test_1, label_train=T_test_1, sample_mean=class_means_1, fsize=(10, 5))

Y_hat_test_1 = nmc_1.classify(X=X_test_1)

CER_test_1 = nmc_1.calculateCER(T=T_test_1, Y_hat=Y_hat_test_1, percentageFlag=True)

accuracy_test_1 = nmc_1.calculateAccuracy(T=T_test_1, Y_hat=Y_hat_test_1, percentageFlag=True)

print(f"Classification Error Rate for the testing set of dataset-1 is: {CER_test_1}")
print(f"Accuracy for the testing set of dataset-1 is: {accuracy_test_1}")
```

## d] Feature transformation: feature reduction by projection

```python
rm = np.zeros((40, 2))
```

```python
for m in range(40):
    if m < 10:
        rm[m] = [10, m]
    elif m >=10 and m < 30:
        rm[m] = [20-m, 10]
    else:
        rm[m] = [-10, 40-m]
```

```python

nmc_1_std = nearestMeansClassifier.NearestMeansClassifier()

X_train_1, n_train_1, T_train_1 = nmc_1_std.generateData(data=train_data_1, printFlag=True)

X_train_1_standardized = nmc_1_std.standardizeData(X=X_train_1, printFlag = True)

CER_History = []
accuracy_History = []
m = []
count = 0
for i, vec in enumerate(rm):

    X_train_1_reduced = np.dot(X_train_1_standardized, vec.reshape(2, 1))


    class_means_1_reduced = nmc_1_std.calculateClassMeans(X=X_train_1_reduced, redFlag=True)

    print("---------------------------------------------------")
    print(f"  Shape of sample_means for {i, vec}: {class_means_1_reduced.shape}")
    print(f"  Sample Means for {vec}: ")
    print(class_means_1_reduced)
    print("---------------------------------------------------")


    Y_hat_train_1_reduced = nmc_1_std.classify(X=X_train_1_reduced)

    CER_train_1_reduced = nmc_1_std.calculateCER(T=T_train_1, Y_hat=Y_hat_train_1_reduced, percentageFlag=True)
    CER_History.append(CER_train_1_reduced )

    accuracy_train_1_reduced = nmc_1_std.calculateAccuracy(T=T_train_1, Y_hat=Y_hat_train_1_reduced, percentageFlag=True)
    accuracy_History.append(accuracy_train_1_reduced)
    
    print(f"Classification Error Rate for the set produced by {i, vec} for dataset-1 is: {CER_train_1_reduced}")
    print(f"Accuracy for the set produced by {i, vec} for dataset-1 is: {accuracy_train_1_reduced}")
    
    m.append(count)
    count += 1
```

```python
plt.plot(m, CER_History)
plt.xlabel('m')
plt.ylabel('Classification Error Rate')
plt.title('Classification Error Rate vs. m')
plt.show()
```

```python
ax = plt.axes()
ax.plot(m, CER_History, c = "grey")


for i in range(40):
    if CER_History[i] == 0.0:
        ax.scatter(i, CER_History[i], c='green', marker='o', s = 50, alpha=1)
    else:
        ax.scatter(i, CER_History[i], c='red', marker='o', s = 50, alpha=1)

ax.set_title("Classification Error Rate vs. m")
ax.set_ylabel('Classification Error Rate (CER)')
ax.set_xlabel('m')


plt.show()
```

```python
ax = plt.axes()
ax.plot(m, accuracy_History, c = "grey")


for i in range(40):
    if accuracy_History[i] == 100.0:
        ax.scatter(i, accuracy_History[i], c='purple', marker='o', s = 50, alpha=1)
    else:
        ax.scatter(i, accuracy_History[i], c='orange', marker='o', s = 50, alpha=1)

ax.set_title("Accuracy vs. m")
ax.set_ylabel('Accuracy')
ax.set_xlabel('m')


plt.show()
```
